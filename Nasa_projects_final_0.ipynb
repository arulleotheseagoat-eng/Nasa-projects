```python
import requests

API_KEY="M0oV1ghvmKnMqhS6c5ynHBzffBewcNx16KLNocYo" # ðŸ” Replace with your actual key
target = 10000
asteroids_data = []

# Starting URL
url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date=2024-01-01&end_date=2024-01-07&api_key={API_KEY}"

while len(asteroids_data) < target:
    response = requests.get(url)
    data = response.json()

    neo_dates = data['near_earth_objects']

    for date, ast_list in neo_dates.items():
        for ast in ast_list:
            for approach in ast['close_approach_data']:
                # Append cleaned asteroid data
                asteroids_data.append(dict(
                    id=ast['id'],
                    neo_id = ast['neo_reference_id'],
                    name=ast['name'],
                    magnitude=ast['absolute_magnitude_h'],
                    dia_min=ast['estimated_diameter']['kilometers']['estimated_diameter_min'],
                    dia_max=ast['estimated_diameter']['kilometers']['estimated_diameter_max'],
                    hazardous=ast['is_potentially_hazardous_asteroid'],
                    closest_approach_date=approach['close_approach_date'],
                    velocity_kmph=float(approach['relative_velocity']['kilometers_per_hour']),
                    astronomical = ast['close_approach_data'][0]['miss_distance']['astronomical'],
                    miss_distance_km=float(approach['miss_distance']['kilometers']),
                    miss_distance_lunar=float(approach['miss_distance']['lunar']),
                    orbiting_body=approach['orbiting_body'],
                ))

                if len(asteroids_data) >= target:
                    break
            if len(asteroids_data) >= target:
                break
        if len(asteroids_data) >= target:
            break

    # ðŸ” Go to the next page (NASA API provides this!)
    if 'next' in data['links']:
        url = data['links']['next']
    else:
        break  # No more data to fetch

print(f"âœ… Total asteroids collected: {len(asteroids_data)}")
```

1.Count how many times each asteroid has approached Earth

```python
cursor.execute("""
SELECT ca.neo_reference_id AS asteroid_id,
       a.name,
       COUNT(*) AS approach_count
FROM close_approach ca
JOIN asteroids a ON ca.neo_reference_id = a.id
GROUP BY ca.neo_reference_id
ORDER BY approach_count DESC
""")
result = cursor.fetchall()
data1 = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data1
```

2) Average velocity of each asteroid (multiple approaches)

```python
cursor.execute("""
SELECT ca.neo_reference_id AS asteroid_id,
       a.name,
       AVG(CAST(ca.relative_velocity_kmph AS REAL)) AS avg_velocity_kmph,
       COUNT(*) AS approaches
FROM close_approach ca
JOIN asteroids a ON ca.neo_reference_id = a.id
GROUP BY ca.neo_reference_id
ORDER BY avg_velocity_kmph DESC
""")
result = cursor.fetchall()
data2 = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data2
```

3) Top 10 fastest asteroids

```python
cursor.execute("""
SELECT a.name,
       MAX(CAST(ca.relative_velocity_kmph AS REAL)) AS max_velocity_kmph
FROM close_approach ca
JOIN asteroids a ON ca.neo_reference_id = a.id
GROUP BY ca.neo_reference_id
ORDER BY max_velocity_kmph DESC
LIMIT 10
""")
result = cursor.fetchall()
data3 = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data3
```

4) Hazardous asteroids that approached Earth more than 3 times

```python
cursor.execute("""
SELECT a.name,
       COUNT(*) AS approach_count
FROM asteroids a
JOIN close_approach ca ON ca.neo_reference_id = a.id
WHERE a.is_potentially_hazardous_asteroid = 1
GROUP BY a.id
HAVING COUNT(*) > 3
ORDER BY approach_count DESC
""")
result = cursor.fetchall()
data4 = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data4
```

5) Month with most asteroid approaches

```python
cursor.execute("""
SELECT substr(close_approach_date,1,7) AS year_month,
       COUNT(*) AS approaches
FROM close_approach
GROUP BY year_month
ORDER BY approaches DESC
LIMIT 1
""")
result = cursor.fetchall()
data5 = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data5
```

âœ… 6) Asteroid with the fastest approach (single event)

```python
cursor.execute("""
SELECT a.name,
       ca.close_approach_date,
       CAST(ca.relative_velocity_kmph AS REAL) AS velocity_kmph
FROM close_approach ca
JOIN asteroids a ON ca.neo_reference_id = a.id
ORDER BY velocity_kmph DESC
LIMIT 1
""")
result = cursor.fetchall()
data6 = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data6
```

âœ… 7) Sort asteroids by max estimated diameter

```python
cursor.execute("""
SELECT name,
       CAST(estimated_diameter_max_km AS REAL) AS max_diameter
FROM asteroids
ORDER BY max_diameter DESC
""")
result = cursor.fetchall()
data7 = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data7
```

âœ… 8) Asteroids whose closest approach is getting nearer over time

```python
cursor.execute("""
WITH earliest AS (
  SELECT neo_reference_id,
         MIN(CAST(miss_distance_km AS REAL)) AS min_miss
  FROM close_approach
  GROUP BY neo_reference_id
),
latest AS (
  SELECT neo_reference_id,
         MAX(CAST(miss_distance_km AS REAL)) AS max_miss
  FROM close_approach
  GROUP BY neo_reference_id
)
SELECT a.name,
       e.min_miss AS earliest_distance,
       l.max_miss AS latest_distance
FROM earliest e
JOIN latest l ON e.neo_reference_id = l.neo_reference_id
JOIN asteroids a ON a.id = e.neo_reference_id
WHERE latest_distance < earliest_distance
""")
result = cursor.fetchall()
data8 = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data8
```

âœ… 9) Closest approach date + distance for each asteroid

```python
cursor.execute("""
WITH min_miss AS (
  SELECT neo_reference_id,
         MIN(CAST(miss_distance_km AS REAL)) AS min_distance
  FROM close_approach
  GROUP BY neo_reference_id
)
SELECT a.name,
       ca.close_approach_date,
       CAST(ca.miss_distance_km AS REAL) AS miss_km
FROM min_miss m
JOIN close_approach ca
     ON ca.neo_reference_id = m.neo_reference_id
    AND CAST(ca.miss_distance_km AS REAL) = m.min_distance
JOIN asteroids a ON a.id = m.neo_reference_id
ORDER BY miss_km ASC
""")
result = cursor.fetchall()
data9 = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data9
```

âœ… 10) Asteroids with velocity > 50,000 km/h

```python
cursor.execute("""
SELECT DISTINCT a.name,
       CAST(ca.relative_velocity_kmph AS REAL) AS velocity_kmph,
       ca.close_approach_date
FROM close_approach ca
JOIN asteroids a ON ca.neo_reference_id = a.id
WHERE velocity_kmph > 50000
ORDER BY velocity_kmph DESC
""")
result = cursor.fetchall()
data10 = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data10
```

âœ… 11) Count how many approaches per month

âœ… 13) Hazardous vs non-hazardous counts

```python
cursor.execute("""
SELECT
  CASE WHEN is_potentially_hazardous_asteroid = 1 THEN 'Hazardous'
       ELSE 'Non-Hazardous' END AS hazard_type,
  COUNT(*) AS count
FROM asteroids
GROUP BY hazard_type
""")
result = cursor.fetchall()
data13 = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data13
```

14) Asteroids that passed closer than 1 Lunar Distance

```python
cursor.execute("""
SELECT a.name,
       ca.close_approach_date,
       CAST(ca.miss_distance_lunar AS REAL) AS miss_ld
FROM close_approach ca
JOIN asteroids a ON ca.neo_reference_id = a.id
WHERE miss_ld < 1
ORDER BY miss_ld ASC
""")
result = cursor.fetchall()
data14 = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data14
```

âœ… 15) Asteroids within 0.05 AU

```python
cursor.execute("""
SELECT a.name,
       ca.close_approach_date,
       CAST(ca.miss_distance_km AS REAL) AS miss_km
FROM close_approach ca
JOIN asteroids a ON ca.neo_reference_id = a.id
WHERE miss_km < (0.05 * 149597870.7)  -- 0.05 AU in km
ORDER BY miss_km ASC
""")

result = cursor.fetchall()
data = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data
```

â­ ADDITIONAL SQL QUERY 5 â€” Asteroids that approached Earth multiple times in last 2 years

```python
cursor.execute("""
SELECT a.name,
       COUNT(*) AS approaches_last_2_years
FROM close_approach ca
JOIN asteroids a ON ca.neo_reference_id = a.id
WHERE close_approach_date >= date('now', '-2 years')
GROUP BY a.name
HAVING approaches_last_2_years >= 2
ORDER BY approaches_last_2_years DESC
""")

result = cursor.fetchall()
data = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data
```

```python
%%writefile app.py

import streamlit as st
import pandas as pd
from streamlit_option_menu import option_menu
from datetime import datetime
import sqlite3

# ------------------ DATABASE CONNECTION ------------------
conn = sqlite3.connect('/content/example.db')
cursor = conn.cursor()

# ------------------ PAGE CONFIG ------------------
st.set_page_config(layout="wide")
st.markdown(
    "<h1 style='text-align: center; color: #4B8BBE;'>ðŸš€ NASA Asteroid Tracker ðŸŒ </h1>",
    unsafe_allow_html=True
)
st.divider()

# ------------------ SIDEBAR ------------------
with st.sidebar:
    selected = option_menu(
        menu_title="Asteroid Approaches",
        options=["Filter Criteria", "Queries"],
        icons=["calendar", "calendar3"],
        menu_icon="cast",
        default_index=0
    )

# =========================================================
# FILTER CRITERIA SECTION
# =========================================================
if selected == "Filter Criteria":

    c1, a, c2, b, c3 = st.columns([0.2, 0.1, 0.2, 0.1, 0.2])

    with c1:
        mag_min = st.slider("Min Magnitude", 13.8, 32.61, (13.8, 32.61))
        diam_min = st.slider("Min Estimated Diameter (km)", 0.00, 4.62, (0.00, 4.62))
        diam_max = st.slider("Max Estimated Diameter (km)", 0.00, 10.33, (0.00, 10.33))

    with c2:
        velocity = st.slider(
            "Relative Velocity (km/h)",
            1418.21, 173071.83,
            (1418.21, 173071.83)
        )
        hazardous = st.selectbox(
            "Only Show Potentially Hazardous",
            options=[0, 1],
            index=0
        )

    with c3:
        start_date = st.date_input("Start Date", datetime(2024, 1, 1))
        end_date = st.date_input("End Date", datetime(2025, 4, 13))

    button = st.button("Filter")

    if button:
        query = """
        SELECT
            a.id, a.name, a.absolute_magnitude_h,
            a.estimated_diameter_min_km, a.estimated_diameter_max_km,
            a.is_potentially_hazardous_asteroid,
            ca.close_approach_date,
            ca.relative_velocity_kmph,
            ca.miss_distance_km,
            ca.orbiting_body
        FROM asteroids a
        JOIN close_approach ca
            ON a.id = ca.neo_reference_id
        WHERE a.absolute_magnitude_h BETWEEN ? AND ?
          AND a.estimated_diameter_min_km BETWEEN ? AND ?
          AND a.estimated_diameter_max_km BETWEEN ? AND ?
          AND ca.relative_velocity_kmph BETWEEN ? AND ?
          AND ca.close_approach_date BETWEEN ? AND ?
          AND a.is_potentially_hazardous_asteroid = ?
        """

        params = [
            mag_min[0], mag_min[1],
            diam_min[0], diam_min[1],
            diam_max[0], diam_max[1],
            velocity[0], velocity[1],
            start_date, end_date,
            hazardous
        ]

        cursor.execute(query, params)
        df = pd.DataFrame(
            cursor.fetchall(),
            columns=[i[0] for i in cursor.description]
        )

        st.subheader("Filtered Asteroids")
        st.dataframe(df, use_container_width=True)

# =========================================================
# QUERIES SECTION
# =========================================================
elif selected == "Queries":

    option = st.selectbox(
        "Select your query",
        [
            "Count how many times each asteroid has approached Earth",
            "Average velocity of each asteroid over multiple approaches",
            "List top 10 fastest asteroids",
            "Find potentially hazardous asteroids that have approached Earth more than 3 times",
            "Find the month with the most asteroid approaches",
            "Get the asteroid with the fastest ever approach speed",
            "Sort asteroids by maximum estimated diameter (descending)",
            "Asteroid whose closest approach is getting nearer over time",
            "Display name + date + miss distance of closest approach",
            "Asteroids that approached Earth with velocity > 50,000 km/h",
            "Count how many approaches happened per month",
            "Find asteroid with highest brightness (lowest magnitude value)",
            "Number of hazardous vs non-hazardous asteroids",
            "Asteroids that passed closer than the Moon (< 1 LD)",
            "Asteroids within 0.05 AU",
            "Asteroids that approached multiple times in last 2 years",
            "Fastest hazardous asteroids",
            "Top 10 asteroids with closest AVERAGE approach distance",
            "Year with the most asteroid approaches",
        ],
        index=None
    )

    if option:

        if option == "Count how many times each asteroid has approached Earth":
            sql = """
            SELECT a.name, COUNT(*) AS approach_count
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_reference_id = a.id
            GROUP BY a.id, a.name
            ORDER BY approach_count DESC
            """

        elif option == "Average velocity of each asteroid over multiple approaches":
            sql = """
            SELECT a.name,
                   AVG(CAST(ca.relative_velocity_kmph AS REAL)) AS avg_velocity_kmph,
                   COUNT(*) AS approaches
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_reference_id = a.id
            GROUP BY a.id, a.name
            ORDER BY avg_velocity_kmph DESC
            """

        elif option == "List top 10 fastest asteroids":
            sql = """
            SELECT a.name,
                   MAX(CAST(ca.relative_velocity_kmph AS REAL)) AS max_velocity
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_reference_id = a.id
            GROUP BY a.id, a.name
            ORDER BY max_velocity DESC
            LIMIT 10
            """

        elif option == "Find potentially hazardous asteroids that have approached Earth more than 3 times":
            sql = """
            SELECT a.name, COUNT(*) AS approach_count
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_reference_id = a.id
            WHERE a.is_potentially_hazardous_asteroid = 1
            GROUP BY a.id, a.name
            HAVING COUNT(*) > 3
            ORDER BY approach_count DESC
            """

        elif option == "Find the month with the most asteroid approaches":
            sql = """
            SELECT substr(close_approach_date,1,7) AS year_month,
                   COUNT(*) AS approaches
            FROM close_approach
            GROUP BY year_month
            ORDER BY approaches DESC
            LIMIT 1
            """

        elif option == "Get the asteroid with the fastest ever approach speed":
            sql = """
            SELECT a.name,
                   ca.close_approach_date,
                   CAST(ca.relative_velocity_kmph AS REAL) AS velocity_kmph
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_reference_id = a.id
            ORDER BY velocity_kmph DESC
            LIMIT 1
            """

        elif option == "Sort asteroids by maximum estimated diameter (descending)":
            sql = """
            SELECT name,
                   CAST(estimated_diameter_max_km AS REAL) AS max_diameter
            FROM asteroids
            ORDER BY max_diameter DESC
            """

        elif option == "Asteroid whose closest approach is getting nearer over time":
            sql = """
            WITH dist AS (
                SELECT neo_reference_id,
                       MIN(CAST(miss_distance_km AS REAL)) AS min_dist,
                       MAX(CAST(miss_distance_km AS REAL)) AS max_dist
                FROM close_approach
                GROUP BY neo_reference_id
            )
            SELECT a.name, d.min_dist, d.max_dist
            FROM dist d
            JOIN asteroids a ON a.id = d.neo_reference_id
            WHERE d.max_dist < d.min_dist
            """

        elif option == "Display name + date + miss distance of closest approach":
            sql = """
            WITH min_miss AS (
                SELECT neo_reference_id,
                       MIN(CAST(miss_distance_km AS REAL)) AS min_distance
                FROM close_approach
                GROUP BY neo_reference_id
            )
            SELECT a.name,
                   ca.close_approach_date,
                   CAST(ca.miss_distance_km AS REAL) AS miss_km
            FROM min_miss m
            JOIN close_approach ca
              ON ca.neo_reference_id = m.neo_reference_id
             AND CAST(ca.miss_distance_km AS REAL) = m.min_distance
            JOIN asteroids a ON a.id = m.neo_reference_id
            ORDER BY miss_km ASC
            """

        elif option == "Asteroids that approached Earth with velocity > 50,000 km/h":
            sql = """
            SELECT a.name,
                   CAST(ca.relative_velocity_kmph AS REAL) AS velocity_kmph,
                   ca.close_approach_date
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_reference_id = a.id
            WHERE CAST(ca.relative_velocity_kmph AS REAL) > 50000
            ORDER BY velocity_kmph DESC
            """

        elif option == "Count how many approaches happened per month":
            sql = """
            SELECT substr(close_approach_date,1,7) AS year_month,
                   COUNT(*) AS total_approaches
            FROM close_approach
            GROUP BY year_month
            ORDER BY year_month
            """

        elif option == "Find asteroid with highest brightness (lowest magnitude value)":
            sql = """
            SELECT name, absolute_magnitude_h
            FROM asteroids
            ORDER BY CAST(absolute_magnitude_h AS REAL) ASC
            LIMIT 1
            """

        elif option == "Number of hazardous vs non-hazardous asteroids":
            sql = """
            SELECT
              CASE WHEN is_potentially_hazardous_asteroid = 1
                   THEN 'Hazardous' ELSE 'Non-Hazardous' END AS hazard_type,
              COUNT(*) AS count
            FROM asteroids
            GROUP BY hazard_type
            """

        elif option == "Asteroids that passed closer than the Moon (< 1 LD)":
            sql = """
            SELECT a.name,
                   ca.close_approach_date,
                   CAST(ca.miss_distance_lunar AS REAL) AS miss_ld
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_reference_id = a.id
            WHERE CAST(ca.miss_distance_lunar AS REAL) < 1
            ORDER BY miss_ld ASC
            """

        elif option == "Asteroids within 0.05 AU":
            sql = """
            SELECT a.name,
                   ca.close_approach_date,
                   CAST(ca.miss_distance_km AS REAL) AS miss_km
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_reference_id = a.id
            WHERE CAST(ca.miss_distance_km AS REAL) < (0.05 * 149597870.7)
            ORDER BY miss_km ASC
            """

        elif option == "Asteroids that approached multiple times in last 2 years":
            sql = """
            SELECT a.name,
                   COUNT(*) AS approaches_last_2_years
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_reference_id = a.id
            WHERE ca.close_approach_date >= date('now', '-2 years')
            GROUP BY a.id, a.name
            HAVING COUNT(*) >= 2
            ORDER BY approaches_last_2_years DESC
            """

        elif option == "Fastest hazardous asteroids":
            sql = """
            SELECT a.name,
                   CAST(ca.relative_velocity_kmph AS REAL) AS velocity_kmph,
                   ca.close_approach_date
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_reference_id = a.id
            WHERE a.is_potentially_hazardous_asteroid = 1
            ORDER BY velocity_kmph DESC
            LIMIT 20
            """

        elif option == "Top 10 asteroids with closest AVERAGE approach distance":
            sql = """
            SELECT a.name,
                   AVG(CAST(ca.miss_distance_km AS REAL)) AS avg_miss_km,
                   COUNT(*) AS approaches
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_reference_id = a.id
            GROUP BY a.id, a.name
            HAVING COUNT(*) >= 2
            ORDER BY avg_miss_km ASC
            LIMIT 10
            """

        elif option == "Year with the most asteroid approaches":
            sql = """
            SELECT strftime('%Y', close_approach_date) AS year,
                   COUNT(*) AS total_approaches
            FROM close_approach
            GROUP BY year
            ORDER BY total_approaches DESC
            LIMIT 1
            """

        cursor.execute(sql)
        df = pd.DataFrame(cursor.fetchall(), columns=[i[0] for i in cursor.description])
        st.dataframe(df, use_container_width=True)
```
